```python
# id: This is a unique identifier for each Airbnb listing. It's often a combination of letters and numbers.

# name: The name or title of the Airbnb listing, which is usually provided by the host.

# host_id: Similar to the listing ID, this is a unique identifier for each host on Airbnb.

# host_name: The name of the host who owns the Airbnb listing.

# neighbourhood_group: In some datasets, Airbnb listings are grouped by larger geographical areas, such as boroughs or districts within a city.

# neighbourhood: The specific neighborhood or area where the Airbnb listing is located.

# latitude and longitude: The geographical coordinates of the Airbnb listing, which can be used to plot the location on a map.

# room_type: The type of room or accommodation offered in the listing (e.g., entire home/apartment, private room, shared room).

# price: The price per night for renting the Airbnb listing.

# minimum_nights and maximum_nights: The minimum and maximum number of nights that a guest can book the listing for.

# number_of_reviews: The total number of reviews that the Airbnb listing has received from guests.

# last_review: The date of the last review for the Airbnb listing.

# reviews_per_month: The average number of reviews that the listing receives per month.

# availability_365: The number of days in a year that the listing is available for booking.

# calculated_host_listings_count: The number of listings that the host has on Airbnb.
```

```python
# outliers

# 75th percentile
seventy_fifth=df["price"].quantile(0.75)

# 25th percentile
twenty_fifth=df["price"].quantile(0.25)

# interquartile range
price_iqr=seventy_fifth-twenty_fifth

# upper threshold
upper=seventy_fifth+(1.5*price_iqr)

# lower threshold
lower=twenty_fifth-(1.5*price_iqr)

print(upper , lower)
```

```python
df.columns
```

```python
df[(df["price"] < lower )| (df["price"] > upper) ] \
[["neighbourhood","room_type","number_of_reviews"]]
```

```python
# Dropping outliers

no_outliers=df[(df["price"] > lower) & (df["price"] < upper)]
```
